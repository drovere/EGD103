## Sorting Algorithms
There are many different ways you could approach sorting numbers from smallest to largest. One example, the selection sort, was covered in the lectures.

Another simple sorting algorithm is the bubble sort. The bubble sort approach looks at pairs of adjacent numbers (ie. bubbles) and swaps them if they are in the wrong order. An example is shown below (source - https://en.wikipedia.org/wiki/Bubble_sort).

First pass - look at pairs of adjacent numbers. Swap them if they are in the wrong order.
* ( **5 1** 4 2 8 ) → ( 1 5 4 2 8 ), Here, algorithm compares the first two elements, and swaps since 5 > 1.
* ( 1 **5 4** 2 8 ) → ( 1 4 5 2 8 ), Swap since 5 > 4
* ( 1 4 **5 2** 8 ) → ( 1 4 2 5 8 ), Swap since 5 > 2
* ( 1 4 2 **5 8** ) → ( 1 4 2 5 8 ), Now, since these elements are already in order (8 > 5), algorithm does not swap them.

Second pass - repeat the same process as the first pass again.
* ( **1 4** 2 5 8 ) → ( 1 4 2 5 8 )
* ( 1 **4 2** 5 8 ) → ( 1 2 4 5 8 ), Swap since 4 > 2
* ( 1 2 **4 5** 8 ) → ( 1 2 4 5 8 )
* ( 1 2 4 **5 8** ) → ( 1 2 4 5 8 )

Now, the array is already sorted, but the algorithm does not know if it is completed. The algorithm needs one additional whole pass without any swap to know it is sorted.

Third pass - repeat the same process again. If no swaps occur, then we know sorting is complete.
* ( **1 2** 4 5 8 ) → ( 1 2 4 5 8 )
* ( 1 **2 4** 5 8 ) → ( 1 2 4 5 8 )
* ( 1 2 **4 5** 8 ) → ( 1 2 4 5 8 )
* ( 1 2 4 **5 8** ) → ( 1 2 4 5 8 )
