## Predicates (Boolean-valued Functions)
Below you will create some predicates, which are functions that output a Boolean: True or False. While you can solve these with if statements, it is even more efficient to solve them with just a Boolean expression.

For example, consider a function named is_positive that outputs True if a value is positive and False otherwise. An intuitive way to solve is written below:

```python
def is_positive(number):
    if number > 0:
        return True
    else:
        return False
```

If we break down what is happening here, we discover that this approach is inefficient. It states to return True if the condition is True and to return False if the condition is False. A more efficient way to write this function is to just return the condition, like below.

```python
def is_positive(number):
    return number > 0:
```
