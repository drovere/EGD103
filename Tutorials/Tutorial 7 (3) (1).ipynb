### Elemetwise functions
Elementwise functionality with arrays isn't just limited to mathematical operators. NumPy contains mathematical functions (eg. sin, cos, exp) that have elementwise functionality.

Suppose you wanted to find the cosine of 100 evenly spaced values between $0$ and $2\pi$. A convenient way to create an evenly spaced array like this is with the linspace function.

````python
# creating a numpy array of 100 evenly spaced values between 0 and 2*pi
x = np.linspace(0, 2*np.pi, 100)
````

````python
# try running the code below with the math module. You should find an exception is raised, since it only works for numeric data types (ie. a single number).
import math
cos_x = math.cos(x)
````

````python
# try running with numpy. It should work :)
cos_x = np.cos(x)
````

A nice way to check whether the cos function has been applied correctly is to visualise the numbers we have computed. This can be done by running the matplotlib code given below.

````python
# run this code to visualise
fig, ax = plt.subplots()
ax.plot(x, cos_x)
ax.set_title('Plot of $y = cos(x)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
````
