## Strings
Strings are a sequence of characters used to form a piece of text. You can create a string with single quotes 'example string' or double quotes "another example string". A multi-line quote is created by using three quotes.
* Since they are a sequence type, you can apply sequence operations, index, slice, and iterate through each character with a for loop
* String methods allow for text to be processed easily
* Can use the input function to ask the user for an input
* Can use a formatted string to clearly communicate messages to the user

More details can be found here: https://docs.python.org/3/library/stdtypes.html#text-sequence-type-str

```python
# strings are a sequence
# support indexing, slicing, iterating, etc.
my_name = 'David Rovere'
```

```python
# indexing example
my_name[0]
```

```python
# slicing example
my_name[-4:]
```

```python
# iterating example
for letter in my_name:
    print(letter)
```

```python
# sequence operator example
my_name * 5
```

```python
# sequence function example
len(my_name)
```

```python
# strings are immutable (can't be modified)
my_name[0] = 'd'
```

```python
# but can create new strings
my_name = 'd' + my_name[1:]
my_name
```

```python
# requesting user input
# will return the value given as a string
my_age = input('How old are you?')
```

```python
# can communicate output clearly with formatted string
my_age_next_year = int(my_age) + 1
f'My age next year will be {my_age_next_year}.' 
```

```python
# there are heaps of string methods that are useful for solving text based problems
# get each word in the string
my_name.split()
```

```python
for word in my_name.split():
    print(len(word))
```

```python
# change case
my_name.upper()
```

```python
example_text = 'I am a large piece of text. Hear me roar.'
words = example_text.split()
cleaned_words = []
for word in words:
    clean_word = word.lower()
    cleaned_words.append(clean_word.strip("."))
cleaned_words
```

## Sets
A set is an unordered collection of data where each value is unique. Sets are mutable, so you can change the values stored in a set. A set can be created with curly braces eg. {1, 2, 3} or with the set function.
* Cannot index since a set is unordered
* Can iterate through a set with a for loop
* Offer a convenient way to remove duplicate entries from a list - just convert it to a set!
* Give access to set operations like intersection and union

More details can be found here: https://docs.python.org/3/library/stdtypes.html#set-types-set-frozenset

```python
# create a set with curly braces
icecream_flavours = {'vanilla', 'chocolate', 'strawberry'}
icecream_flavours
```

```python
# sets are unordered - can't index or slice
icecream_flavours[0]
```

```python
# all values in a set are unique - cannot contain duplicate entries
fruits = {'apple', 'banana', 'apple', 'strawberry'}
fruits
```

```python
# convert to a set
numbers = [1, 2, 3, 4, 5]
set(numbers)
```

```python
# create an empty
empty_set = set()
empty_set
```

```python
# have access to set operations
# intersection (and)
icecream_flavours & fruits
```

```python
# union (or)
icecream_flavours | fruits
```

```python
# sets are mutable
icecream_flavours.add('caramel')
icecream_flavours
```

## Dictionaries
A dictionary is a set of key : value pairs. The key serves as some unique identifier, and the value provides some information associated with that key. You can create a dictionary with curly braces.
* Dictionaries use a named index (the key name) rather than a positional index.
* Ordered based on when entry was added to dictionary.
* Can use the keys method to access dictionary keys, the values method to access dictionary values, and the items method to access both.
* Can iterate through the keys or values (or both) with a for loop.
* A good way to represent data or physical objects.

More details can be found here: https://docs.python.org/3/library/stdtypes.html#mapping-types-dict

```python
# examples
items = {'potion': 2, 'pokeball': 6, 'antidote': 1}
items
```

```python
# selecting a value using its key
items['pokeball']
```

```python
# dictionaries are mutable
# create a new entry
items['masterball'] = 1
items
```

```python
# modify an existing entry
items['potion'] = 4
items
```

```python
# dictionary methods are useful!
# keys
items.keys()
```

```python
# values
items.values()
```

```python
# items
items.items()
```

```python
# iterating with a for loop
for name, quantity in items.items():
    print(f'You have {quantity} instances of {name} in your bag.')
```

### Activity: Account Login
Design a simple program that gets someone to log in to their account. It must have the following functionality:
* All usernames and passwords that exist are stored as key: value pairs in a dictionary.
* The function should ask the user for their username and password.
* The function should give the user a maximum of 3 incorrect login attempts before locking their account.

```python
# create a dictionary that contains some username: password pairs.
login_details = {'my_account': '1234',
                'top_fan': 'letmein',
                'magic_man': 'open_sesame'}
```

```python
# define a function that gives the user 3 attempts to log in
def account_login():

    n_attempts_remaining = 3
    while n_attempts_remaining > 0:
        username = input('What is your username?')
        password = input('What is your password?')
        if username not in login_details or password != login_details[username]:
            n_attempts_remaining = n_attempts_remaining - 1
            print(f'Incorrect username or password. You have {n_attempts_remaining} incorrect attempts remaining.')
        else:
            print('Login successful')
            return
    print('You have used up all login attempts. Account will be locked for 24 hours.')
                  
        
```

```python
# test function
account_login()
```
