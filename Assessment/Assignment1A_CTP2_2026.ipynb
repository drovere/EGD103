## A1: An Introduction to Crank-Slider Mechanisms
Crank slider mechanisms are used to convert between rotational motion and reciprocating linear motion. Some example applications include:
1. In internal combustion engines, expanding gas in a cylinder moves the piston (slider), which in turn allows the crankshaft to rotate, causing the vehicle to move.
2. In pumps and compressors a motor drives the crank, which then causes the piston (slider) to move backwards and forwards to compress air.

## A5: Classifying Crank Slider
The type of motion you get from a crank-slider mechanism can be determined as a simple function of the crank length ($a$), coupler length ($b$) and slider offset ($c$).
* Mechanism is continuous if $a + c < b$. This means the crank can rotate a full 360 degrees.
* Mechanism is a structure if $a + b = c$. This means it won't be able to move at all.
* Mechanism is not possible if $c > a + b$. 
* For all other conditions, the mechanism is a rocker. This means the crank rocks back and forth rather than being able to rotate 360 degrees.
