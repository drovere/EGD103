## 1.1 Vector Addition
Vector addition is an element-wise process defined by $$\vec{v_1} + \vec{v_2} = \begin{pmatrix}x_1\\y_1\end{pmatrix} + \begin{pmatrix}x_2\\y_2\end{pmatrix} = \begin{pmatrix}x_1 + x_2\\y_1 + y_2\end{pmatrix}$$
The function below shows you how this can be implemented for tuples in Python.

```python
# This function has been implemented for you. Do not change it!
def add_vectors(vector1, vector2):
    (x1, y1) = vector1 # extract the x,y coordinates of the first vector
    (x2, y2) = vector2 # extract the x,y coordinates of the second vector
    sum_x = x1 + x2 # add the x-components together
    sum_y = y1 + y2 # add the y-components together
    return (sum_x, sum_y) # return the resulting vector sum
```

## 2.5 Motion
Once the resultant force is known, the motion of a particle can be calculated. 

Firstly, we can compute acceleration through Newton's 2nd Law:

$$\vec a = \frac{\vec F_R}{m}$$

Then the velocity can be solved numerically using the forward Euler method: 

$$\vec v_{new} = \vec v_{old} + \vec a_{old} \times \Delta t$$

Finally, the position can be solved numerically using the backward Euler method:

$$\vec x_{new} = \vec x_{old} + \vec v_{new} \times \Delta t$$

The formulas for velocity and position are only approximate, since they assume constant rate of change over the time step $\Delta t$, but they are relatively accurate so long as the time step is small.

# Part 3: Simulating a Hanging Cable
Now you will be using the toolbox of vector, force and kinematics functions you have created in order to simulate a hanging cable.

We assume that each end of a cable is fixed securely and so cannot move. We simulate the shape of the cable as it hangs between supports
by pretending that the cable is made up of a finite number of particles. Such Finite element analysis (FEA) is commonly applied to create computer models for a huge variety of engineering problems from vibration, heat and fluid flow, car crash simulation and weather modelling. 

<center></center>

If the entire cable has a mass of M and we divide it into N equal length parts/particles, then each particle will have a mass of $M/N$. Other cable properties may similarly require scaling when divided into particles.
Each particle will have a distinct physical location and current velocity. Since the particles of the cable are connected to each other, we model this connection 
as a set of springs which connect each particle to the particles immediately to its left and right:


The simulation needs to start in some initial state, so we initially position all particles, equally distanced apart, in a straight line between the two fixed attachment points. Note that the supports may not necessarily be at the exact same height. We also give each particle an initial velocity of zero (i.e. they are initially artificially suspended and not moving). We then *"drop"* the particles, allowing gravity and the force of the springs to take effect. We also account for drag due to wind resistance which will tend to dampen the bouncing of the cable. The simulation proceeds via a sequence of small simulation steps representing what happens to each particle during a very short time interval. At each time step we recompute forces and update the velocities and positions of all particles. This is referred to as a Discrete Time simulation. Eventually the position of each particle will stabilize revealing he natural hang of the cable. This shape is referred to as *Catenary* and occurs commonly in physics and nature. It is similar but different from a Parabola.

## 4.2 Initial Conditions
Here we determine the initial position of each section of the cable based on the positions of the attached fixed endpoints.

Each point is initially positioined equally distanced from each other, in a straight line between the two fixed endpoints.

The initial velocity of each section of the cable is also set to zero.

```python
# This function has been implemented for you (do not change it!)

# Use linear interpolation to find a point between p1 and p2. 
# posn is assumed to be a fraction between 0 and 1, where posn=0 will give us p1 and posn=1 will give us p2
# and a posn of 0.5 will give us the point half way between p1 and p2
def interpolate(posn, p1, p2) :
    return (posn-1) * p1 + posn * p2

# create global variables to represent the current position and velocity of each part of the cable
# These positions and velocities will change as the simulation proceeds
positions = [] 
velocities = []

# This function places each part of the cable at an initial location (in a straight line between the two fixed attachment points)
# Don't alter this function, it has been provided for you.
# Later we will let gravity and the springs take effect to find how the cable hangs naturally through simulation
def initialize_positions() :
    
    # clear they position of velocity lists of any previous data, so that we can reset them to their initial start positions here
    positions.clear()
    velocities.clear()
    
    # find the x, y coordinates of each end point
    (xa, ya) = endpoint_a
    (xb, yb) = endpoint_b

    for i in range(number_of_particles) :
        # initially position points equally spaced in a straight line between the two fixed attachment end point
        positions.append((interpolate(i/(number_of_particles-1), xa, xb), interpolate(i/(number_of_particles-1), ya, yb)))
        
        # the velocity of all points is initially zero (i.e. they are not moving)
        velocities.append((0, 0))
```
